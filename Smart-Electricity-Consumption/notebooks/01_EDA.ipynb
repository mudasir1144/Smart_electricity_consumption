3.4.3: Counting and detecting Outlier Values

```python
def detect_outliers(dataframe , columns):
    Q1 = dataframe[columns].quantile(0.25)
    Q3 = dataframe[columns].quantile(0.75)

    IQR= Q3-Q1

    lower = Q1-1.5*IQR
    upper = Q3+1.5*IQR

    outliers = dataframe[(dataframe[columns]<lower)| (dataframe[columns]>upper)]
    return len(outliers)
print('Outliers found:')
for col in numerical_cols:
    print(f"{col}:{detect_outliers(dataset,col)}")

```

5.3: Feature Importance

```python
from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor(random_state=42)
rf.fit(X,y)
```

```python
importance =pd.DataFrame({
    "Feature": X.columns,
    "Importance":rf.feature_importances_
})
importance = importance.sort_values(
    by= "Importance",
    ascending=False
)
display(importance)
```

```python
plt.figure(figsize=(6,4))
sn.barplot(
    x= "Importance",
    y= 'Feature',
    data = importance
)
plt.title('Feature Importance')
plt.show()
```

Insights: AC_Hours and Outdoor_Temprature is the most important feature of our dataset
