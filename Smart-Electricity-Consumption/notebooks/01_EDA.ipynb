3.4.2: Visualizaing outliers:

3.4.3: Counting and detecting Outlier Values

```python
def detect_outliers(dataframe , columns):
    Q1 = dataframe[columns].quantile(0.25)
    Q3 = dataframe[columns].quantile(0.75)

    IQR= Q3-Q1

    lower = Q1-1.5*IQR
    upper = Q3+1.5*IQR

    outliers = dataframe[(dataframe[columns]<lower)| (dataframe[columns]>upper)]
    return len(outliers)
print('Outliers found:')
for col in numerical_cols:
    print(f"{col}:{detect_outliers(dataset,col)}")

```

3.5.1: Feature encoding

```python
categorical_cols = dataset.select_dtypes(include=['object']).columns.tolist()
categorical_cols
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for column in categorical_cols:
    dataset[column]=label_encoder.fit_transform(dataset[column])

dataset.head(2)
```

3.5.2: Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X=dataset.drop('Daily_Electricity_Consumption',axis=1)
Y=dataset['Daily_Electricity_Consumption']
X_scaled= scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled , columns = X.columns)
```

3.6.1: Saving processed dataset

```python
processed_data = X_scaled.copy()
processed_data['Daily_Electricity_Consumption'] = Y

processed_data.to_csv('Smart-Electricity-Consumption\\dataset\\processed_data.csv',index=False)
```
