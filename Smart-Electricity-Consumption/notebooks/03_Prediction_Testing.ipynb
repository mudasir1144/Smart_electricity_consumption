1.1 : Importing libraries

```python
import numpy as np
import pandas as pd
import joblib
```

1.2: Loading models using Joblib

```python
model = joblib.load("C:\\Users\\HP\\OneDrive\\Desktop\\ML-03_TC-INT-20260601-328\\Smart-Electricity-Consumption\\models\\best_model.pkl")
scaler = joblib.load('C:\\Users\\HP\\OneDrive\\Desktop\\ML-03_TC-INT-20260601-328\\Smart-Electricity-Consumption\\models\\scaler.pkl')
encoder = joblib.load("C:\\Users\\HP\\OneDrive\\Desktop\\ML-03_TC-INT-20260601-328\\Smart-Electricity-Consumption\\models\\encoder.pkl")
print("Models loaded successfully!")
```

1.3: Creating sample data 

```python
sample_input = {
    "House_Type": "House",
    "Family_Members": 4,
    "Rooms": 5,
    "AC_Hours": 8,
    "Fan_Hours": 10,
    "Refrigerator": 1,
    "Washing_Machine_Hours": 1.2,
    "Water_Motor_Hours": 1.5,
    "Lighting_Hours": 8,
    "Outdoor_Temperature": 35,
    "Season": "Summer",
    "Day_Type": "Weekday",
    "Work_From_Home": "No",
    "Solar_Panels": "No"
}
sample_data = pd.DataFrame([sample_input])
sample_data
```

1.4: Encoding the categorical columns

```python
categorical_columns = [
    "House_Type",
    "Season",
    "Day_Type",
    "Work_From_Home",
    "Solar_Panels"
]

for column in categorical_columns:
    sample_data[column] = encoder[column].transform(sample_data[column])

sample_data
```

1.5: Scaled the data which is encoded

```python
scaled_data = scaler.transform(sample_data)
scaled_data
```

1.6: Predicting from scaled data

```python
prediction = model.predict(scaled_data)
```

1.7: Predicting daily consumption

```python
daily_consumption = round(prediction[0],2)
print(f"Predicted daily consumption : {daily_consumption} KWH")
```

1.8: Predicting monthly consumption

```python
monthly_consumption = round(daily_consumption*30 ,2)
print(f"Monthly consumption : {monthly_consumption} Kwh")
```

1.9: Estimated Monthly Bill

```python
unit_price = 65

monthly_bill = round(monthly_consumption*unit_price , 2)
print(f"Estimared monthly bill will be :{monthly_bill} rs")
```
